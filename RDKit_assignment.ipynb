```python
#import libraries
from rdkit import Chem
from rdkit.Chem import Descriptors, Draw
import pandas as pd
```

```python
#SMILES notation for selected molecules
Propane = 'CCC'
Ethene = 'C=C'
Cyclohexane = 'C1CCCCC1'
Buckminsterfullerene = 'C12=C3C4=C5C6=C1C7=C8C9=C1C%10=C%11C(=C29)C3=C2C3=C4C4=C5C5=C9C6=C7C6=C7C8=C1C1=C8C%10=C%10C%11=C2C2=C3C3=C4C4=C5C5=C%11C%12=C(C6=C95)C7=C1C1=C%12C5=C%11C4=C3C3=C5C(=C81)C%10=C23'
```

```python
# storing SMILES representation of molecules in dictionary
smiles = {'Propane': 'CCC', 'Ethene': 'C=C', 'Cyclohexane': 'C1CCCCC1',
          'Buckminsterfullerene': 'C12=C3C4=C5C6=C1C7=C8C9=C1C%10=C%11C(=C29)C3=C2C3=C4C4=C5C5=C9C6=C7C6=C7C8=C1C1=C8C%10=C%10C%11=C2C2=C3C3=C4C4=C5C5=C%11C%12=C(C6=C95)C7=C1C1=C%12C5=C%11C4=C3C3=C5C(=C81)C%10=C23'}
```

```python
#Create RDKit molecule objects
molecules = {}

for name, smile in smiles.items():
    mol = Chem.MolFromSmiles(smile)
    molecules[name] = mol
```

```python
# rdkit molecule objects
for name, mol in molecules.items():
    print(f'{name}: {Chem.MolToSmiles(mol)}')
```

```python
#number of atoms in each molecule
for name, mol in molecules.items():
    num_atoms = mol.GetNumAtoms()
    print(f'{name} has {num_atoms} atoms.')
```

```python
# printing theh chemical symbol and atomic weight of each atom
for name, mol in molecules.items():
    print(f'{name}:') # print name of molecule
    for atom in mol.GetAtoms():
        symbol = atom.GetSymbol()
        atomic_weight = atom.GetMass()
        print(f'Atom: {symbol}, Weight: {atomic_weight:.2f}')
    print()
```

```python
#number of aromatic bonds in each molecule
for name, mol in molecules.items():
    aromatic_bonds = 0
    for bond in mol.GetBonds():
        if bond.GetIsAromatic():
            aromatic_bonds += 1
    print(f'{name} has {aromatic_bonds} aromatic bonds.')
```

# Challenges Encountered

One of the main challenges encountered during this assignment was related to parsing complex SMILES strings in RDKit, particularly for the molecule Buckminsterfullerene (C60). While simple organic molecules such as propane, ethene and cyclohexane were processed easily, the SMILES representation of Buckminsterfullerene retrieved from PubChem proved problematic. RDKit consistently raised parsing errors such as "unclosed ring," indicating that the SMILES notation exceeded the parser's capabilities in the version of RDKit available on Google Collab. This issue occurs because C60 contains multiple fused rings and double-digit colusers (eg., %10, %11, %12) that older RDKit builds cannot interpret. To overcome this, an alternative method using a fully compatible SMILES string [C12=C3C4=C5C6=C1C7=C8C9=C1C%10=C%11C(=C29)C3=C2C3=C4C4=C5C5=C9C6=C7C6=C7C8=C1C1=C8C%10=C%10C%11=C2C2=C3C3=C4C4=C5C5=C%11C%12=C(C6=C95)C7=C1C1=C%12C5=C%11C4=C3C3=C5C(=C81)C%10=C23] which successfully generated the molecule without error. This confirmed that RDKit is capable of interpreting complex SMILES notations when properly formatted. In addition, occasional NoneType errors occured when valid SMILES strings failed to produce valid molecule objects. These were mitigated through the use of try-except statements, allowing the program to handle such expections gracefully without interrupting execution.

Another challenge arose during the visualization phase, particularly when displayed the first six compounds from Delaney dataset. Initially, compound labels were not correctly rendered in the molecular grid due to mismatched data formats. This issues was resolved by converting the pandas Series of compound identifiers into a list before passing it to visualization function. Once corrected, the molecular grid displayed each compound clearly with its corresponding label, improving both readibility and presentation quality.
